```python
## 環境変数設定
import dotenv
dotenv.load_dotenv('./drive/MyDrive/openai.env')
```

```python
## キーフレーズのリスト
keyphrases = [
    'Python',
    '機械学習',
    '勉強会',
    '新潟',
]
```

```python
## 雰囲気に関わるリスト
atmospheres = [
    'クール',
    'おしゃれ',
    'ほのぼの',
    'あたたかい',
]
```

```python
import random
import os
from openai import OpenAI

import base64
from PIL import Image
from IPython.display import display_png
from io import BytesIO
from google.colab.patches import cv2_imshow  # for image display
from skimage import io

client = OpenAI(
    api_key = os.getenv("OPENAI_API_KEY"),  # set OpenAI API Key
)
cnt = 1
```

```python
phrase = f'{",".join(keyphrases)}, {",".join(random.sample(atmospheres, 2))}, 背景は白'

## Prompt
prompt=f'''\
以下のフレーズを元に、画像の中に「PyML Niigata」という文字を入れた、ロゴ画像を作ってください。ロゴ画像以外は描かないでください。

{phrase}
'''
```

```python
response = client.images.generate(
  model="dall-e-3",
  prompt=prompt,
  size="1024x1024",
  quality="standard",
  n=1,
  response_format="b64_json",
)

# image_url = response.data[0].url
b64 = response.data[0].b64_json
print(prompt)
```

```python
# image_url
b64
```

```python
## 画像表示

# image = io.imread(image_url)
# cv2_imshow(image)

image_data = base64.b64decode(b64)
image_stream = BytesIO(image_data)
image = Image.open(image_stream)
path=f"./drive/MyDrive/pyml_logo_{cnt}.png"
image.save(path)

display_png(Image.open(path))

cnt += 1
```
